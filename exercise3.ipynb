```python
#计算离散型变量的IV,
def calc_iv_object(Y,X):
    dframe = pd.DataFrame({'X':X,'Y':Y})
    dframe.dropna(inplace=True)
    total_bad = dframe[dframe['Y']==1].shape[0]
    total_good =dframe[dframe['Y']==0].shape[0]
    dframe_group_bad = dframe.groupby(['X']).Y.sum().reset_index()
    dframe_group_bad.columns=['X_1','group_bad']
    dframe_group_total = dframe.groupby(['X']).Y.count().reset_index()
    dframe_group_total.columns=['X_2','group_total']
    dframe_merge=pd.merge(dframe_group_bad,dframe_group_total,how='left',left_on=['X_1'],right_on=['X_2'])
    dframe_merge['good']=dframe_merge['group_total']-dframe_merge['group_bad']
    dframe_merge['WOE']=np.log((dframe_merge['group_bad']/total_bad)/(dframe_merge['good']/total_good))
    dframe_merge['IV']=((dframe_merge['group_bad']/total_bad)-(dframe_merge['good']/total_good))*dframe_merge['WOE']
    return np.sum(dframe_merge['IV'].values)
```

```python
#连续性变量先要分箱，这里用等频分箱
def calc_iv_num(Y,X,n=10):
    X2=X.fillna(np.median(X))
    r=0
    while np.abs(r)<1:
        d1=pd.DataFrame({'X':X2,'Y':Y,'bucket':pd.qcut(X2,n,duplicates='drop')})
        d2=d1.groupby('bucket',as_index=True)
        r,p=stats.spearmanr(d2.mean().X,d2.mean().Y)
        n -=1
    total_good =d1[d1['Y']==0].shape[0]
    total_bad = d1[d1['Y']==1].shape[0]
    dframe_group_bad=d1.groupby(['bucket']).Y.sum().reset_index()
    dframe_group_bad.columns=['bucket','bad']
    dframe_group_total=d1.groupby(['bucket']).Y.count().reset_index()
    dframe_group_total.columns=['bucket','total']
    dframe_merge=pd.merge(dframe_group_bad,dframe_group_total,how='left',left_on=['bucket'],right_on=['bucket'])
    dframe_merge['good']=dframe_merge['total']-dframe_merge['bad']
    dframe_merge['WOE']=np.log((dframe_merge['bad']/total_bad)/(dframe_merge['good']/total_good))
    dframe_merge['IV']=((dframe_merge['bad']/total_bad-dframe_merge['good']/total_good))*dframe_merge['WOE']
    return np.sum(dframe_merge['IV'].values)
```
