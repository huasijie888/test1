```python
#数据质量分析，分为每个特征数量，缺失值情况，唯一值个数，唯一率
def desc(frame):
    loss_lst=[]
    total_num=[]
    unique_percent=[]
    unique_num=[]
    for i,j in enumerate(frame.columns):
        loss_lst.append(frame[frame[j].isnull()].shape[0])
        total_num.append(frame[frame[j].notnull()].shape[0])
        unique_percent.append(len(frame[frame[j].notnull()][j].unique())/frame[frame[j].notnull()].shape[0])
        unique_num.append(len(frame[j].unique()))
    data_count=pd.DataFrame({'loss_count':loss_lst,'total_num':total_num,'unique_percent':unique_percent,'unique_num':unique_num},index=frame.columns)
    return data_count
desc_frame=desc(data)
desc_frame
```

```python
#分别分析字符类型和数值类型两种特征
num_feature=data.select_dtypes(include=['float64','int64']).copy()
word_feature=data.select_dtypes(include=['object']).copy()
```

```python
#首先分析字符类型变量
word_feature_desc = desc(word_feature)
word_feature_desc
```

```python
#订单号本事没作用，但是和latest_query_time，loans_latest_time两个字段可以构成时间差（按天数），对字符类型变量做转化
#根据观察本事订单号前几位因该未进件日期
#首先把进件编号前几位取出来，做成datatime类型字段，再和后面的日期计算相差天数，添加到dataframe中
#转换前先去掉Null数据
data.drop(index=list(data[data['latest_query_time'].isnull()].index),axis=0,inplace=True)
data.drop(index=list(data[data['loans_latest_time'].isnull()].index),axis=0,inplace=True)
latest_query_time_days=[]
loans_latest_time_days=[]
for i in word_feature.index:
    a=datetime.datetime.strptime(data['trade_no'][i][0:4]+'-'+data['trade_no'][i][4:6]+'-'+data['trade_no'][i][6:8],'%Y-%m-%d')-datetime.datetime.strptime(data['latest_query_time'][i],'%Y-%m-%d')
    b=datetime.datetime.strptime(data['trade_no'][i][0:4]+'-'+data['trade_no'][i][4:6]+'-'+data['trade_no'][i][6:8],'%Y-%m-%d')-datetime.datetime.strptime(data['loans_latest_time'][i],'%Y-%m-%d')
    latest_query_time_days.append(a.days)
    loans_latest_time_days.append(b.days)
data['latest_query_time_days']=latest_query_time_days
data['loans_latest_time_days']=loans_latest_time_days
```

```python
#删除掉原来特征
data.drop(columns=['latest_query_time','loans_latest_time'],axis=1,inplace=True)
```

```python
#删除订单号
data.drop(columns=['trade_no'],axis=1,inplace=True)
```

```python
#剩下的字符类型字段，id_name无用删除，source，bank_card_no都只有一个值无用删除，只剩一个reg_preference_for_trad有5种情况，用数字代替
data.drop(columns=['id_name','bank_card_no','source'],axis=1,inplace=True)
```
